```python
#Road accident analysis
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Year': [2019, 2020, 2021, 2022, 2023, 2024],
    'Accidents': [4500, 3800, 4000, 4700, 5200, 5600],
    'Deaths': [1200, 1000, 1100, 1300, 1500, 1600],
    'Injured': [3000, 2500, 2700, 3200, 3500, 3800]
}

df = pd.DataFrame(data)
print(df)
```

```python
print(df.describe())
```

```python
plt.figure(figsize=(8,5))
plt.plot(df['Year'], df['Accidents'], marker='o', label='Accidents')
plt.plot(df['Year'], df['Deaths'], marker='o', label='Deaths')
plt.plot(df['Year'], df['Injured'], marker='o', label='Injured')

plt.title("Road Traffic Accidents Trend")
plt.xlabel("Year")
plt.ylabel("Count")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(df['Year'], df['Accidents'], label='Accidents', color='orange')
plt.bar(df['Year'], df['Deaths'], label='Deaths', color='red')

plt.title("Accidents vs Deaths")
plt.xlabel("Year")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(df['Injured'], labels=df['Year'], autopct='%1.1f%%')
plt.title("Injury Distribution by Year")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['Accidents', 'Deaths', 'Injured']].corr(),
            annot=True, cmap='coolwarm')
plt.title("Correlation between Accident Factors")
plt.show()
```
